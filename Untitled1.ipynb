```python
plt.figure(figsize= (7,5))
plt.hist(df['ocean_proximity'], bins= 12, color = "green")
plt.xlabel("ocean proximity")
plt.ylabel("frequency")
plt.title(" Distribution of ocean_proximity")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['households'],df['population'])
plt.xlabel("households")
plt.ylabel("population")
plt.title("Realtionship b/w households and population")
plt.show()
```

using seaborn (scatter plot)


```python
plt.figure(figsize=(8, 6))

sns.scatterplot(x='households', y='population', data=df, color='orange')

plt.title("Relationship between households and population")

plt.xlabel("households")

plt.ylabel("population")

plt.show()
```

```python
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df_numeric = df.drop(columns=non_numeric_cols)

corr_matrix = df_numeric.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'population', data = df)
plt.xlabel("population")
plt.ylabel("count")
plt.title("Count of population ")
plt.show()
```
